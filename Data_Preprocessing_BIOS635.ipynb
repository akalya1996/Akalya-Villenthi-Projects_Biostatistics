```python
#cleaning and preprocessing for train data
# Handle missingness and change the data format and structure for easy computation

import pandas as pd
import matplotlib.pyplot as plt

# Read train and test datasets
df_train = pd.read_csv('BIOS635/Kaggle/train.csv')
df_test = pd.read_csv('BIOS635/Kaggle/test.csv')

# Data exploration and preprocessing for train dataset
df_train.dtypes
df_train.columns
df_train.head()
df_train.tail()
df_train.isna().any()
df_train.info()
df_train.describe(include='all')

(df_train.CryoSleep == True).sum()/len(df_train)

df_train_copy = df_train.copy()

df_train_copy['Expenses'] = df_train_copy[['RoomService', 'FoodCourt',
                                           'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)

df_train_copy.Age = df_train_copy.Age.fillna(df_train_copy.Age.median())
df_train_copy.loc[df_train_copy.Adult_spending_awake ==False, 'Age'] = 0

df_train_copy['Adult_spending_awake'] = (df_train_copy['Expenses'] > 0) & (df_train_copy['Age'] >=13) & (df_train_copy['CryoSleep'] == False)

df_train_copy['Cryosleep'] = 0
df_train_copy.loc[df_train_copy['Expenses'] == 0, 'Cryosleep'] = 1
df_train_copy.loc[df_train_copy.CryoSleep.astype('str') == 'True', 'Cryosleep'] = 1
df_train_copy.loc[df_train_copy.CryoSleep.astype('str') == 'False', 'Cryosleep'] = 0
df_train_copy['Cryosleep'] = df_train_copy['Cryosleep'].astype('bool')
df_train_copy['CryoSleep'] = df_train_copy['Cryosleep']
df_train_copy.drop('Cryosleep',axis=1,inplace=True)
df_train_copy.drop('Name',axis=1,inplace=True)

df_train_copy.loc[df_train_copy.CryoSleep == True,['RoomService', 'FoodCourt','ShoppingMall', 'Spa', 'VRDeck']] = 0
df_train_copy.loc[df_train_copy.CryoSleep == True,['RoomService', 'FoodCourt','ShoppingMall', 'Spa', 'VRDeck']].isna().sum()

df_train_copy['Age'].plot(kind='hist')
df_train_copy['Adults'] = df_train_copy['Age'] >= 13

df_train_copy['Adult_and_spending'] = (df_train_copy['Expenses'] > 0) & (df_train_copy['Age'] >=13)
df_train_copy.loc[df_train_copy.Adult_and_spending == True]

df_train_copy.RoomService = df_train_copy.RoomService.fillna(df_train_copy.RoomService.mean())
df_train_copy.loc[df_train_copy.Adult_and_spending ==False, 'RoomService'] = 0

df_train_copy.FoodCourt = df_train_copy.FoodCourt.fillna(df_train_copy.FoodCourt.mean())
df_train_copy.loc[df_train_copy.Adult_and_spending ==False, 'FoodCourt'] = 0

df_train_copy.ShoppingMall = df_train_copy.ShoppingMall.fillna(df_train_copy.ShoppingMall.mean())
df_train_copy.loc[df_train_copy.Adult_and_spending ==False, 'ShoppingMall'] = 0

df_train_copy.Spa = df_train_copy.Spa.fillna(df_train_copy.Spa.mean())
df_train_copy.loc[df_train_copy.Adult_and_spending ==False, 'Spa'] = 0

df_train_copy.VRDeck = df_train_copy.VRDeck.fillna(df_train_copy.VRDeck.mean())
df_train_copy.loc[df_train_copy.Adult_and_spending ==False, 'VRDeck'] = 0

df_train_copy.HomePlanet = df_train_copy.HomePlanet.fillna('Earth')
df_train_copy.Destination = df_train_copy.Destination.fillna('TRAPPIST-1e')
df_train_copy.VIP = df_train_copy.VIP.fillna('False')
df_train_copy.VIP = df_train_copy.VIP.astype('bool')

df_train_copy['Cabin'] = df_train_copy.Cabin.fillna(method='ffill')

df_train_copy['Group_nums'] = df_train_copy.PassengerId.apply(lambda x: x.split('_')).apply(lambda x: x[0])
df_train_copy['Grouped'] = ((df_train_copy['Group_nums'].value_counts() > 1).reindex(df_train_copy['Group_nums'])).tolist()
df_train_copy['Deck'] = df_train_copy.Cabin.apply(lambda x: str(x).split('/')).apply(lambda x: x[0])
df_train_copy['Side'] = df_train_copy.Cabin.apply(lambda x: str(x).split('/')).apply(lambda x: x[2])
df_train_copy['Has_expenses'] = df_train_copy['Expenses'] > 0
df_train_copy['Is_Embryo'] = df_train_copy['Age'] == 0

plt.bar(df_train_copy['Side'],df_train_copy['Transported'])
plt.bar(df_train_copy['Deck'],df_train_copy['Transported'])

df_train_copy.columns
df_train_copy.drop(['IsEmbryo'],axis=1, inplace=True)

df_train_copy.to_csv('train_data_clean.csv',index=False)

```
